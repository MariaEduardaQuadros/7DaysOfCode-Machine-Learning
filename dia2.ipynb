```python
def missing_visualization(df):
  quant_isnull = df.isnull().sum()
  columns = df.columns
  dic = {"colunas":[],"quant_isnull":[]}
  for coluna,quant in zip(columns,quant_isnull):
    if quant > 0:
      dic["colunas"].append(quant)
      dic["quant_isnull"].append(coluna)
  df = pd.DataFrame(dic)
  plt.figure(figsize=(15,5))
  sns.barplot(x=df["quant_isnull"],y=df["colunas"],data=df, palette="rocket")
  plt.xticks(rotation=45);
```

```python
missing_visualization(dados)
```

**Dia 02**: conceitos iniciais de Machine Learning e Pré-Processamento de Dados

```python
dados = dados.drop_duplicates()
dados.head()
```

```python
dados[dados["popularity"] >=80]

conditionlist = [
    (dados['popularity'] >= 80) ,
    (dados['popularity'] <80)]

choicelist = [1,0]
dados['pop_classe'] = np.select(conditionlist, choicelist, default='Not Specified')
dados['pop_classe'] = dados['pop_classe'].astype(int)
```

```python
dados = dados.dropna() #removendo valores nulos
```

```python
dados.columns
```

```python
dados = dados.drop(columns=['popularity'])
```
